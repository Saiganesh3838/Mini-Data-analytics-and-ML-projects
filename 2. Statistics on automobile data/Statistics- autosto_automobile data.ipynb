### Detecting Outliers

```python
fig, axes = plt.subplots(nrows=2,ncols=3)
fig.set_size_inches(20,10)
sns.boxplot(data=df,x='Age',ax=axes[0][0])
sns.boxplot(data=df,x='No_of_Dependents',ax=axes[0][1])
sns.boxplot(data=df,x='Total_salary',ax=axes[1][0])
sns.boxplot(data=df,x='Salary',ax=axes[1][1])
sns.boxplot(data=df,x='Partner_salary',ax=axes[0][2])
sns.boxplot(data=df,x='Price')
plt.show()
```

```python
### yes there is outliers in no_of_dependents and total_salary
```

### Outliers treatment

```python
sns.boxplot(data=df,x='No_of_Dependents')
plt.show()
sns.boxplot(data=df,x='Total_salary')
```

```python
min_No_of_Dependents, max_No_of_Dependents = df['No_of_Dependents'].quantile([0.05,0.95])
print(min_No_of_Dependents,2)
print(max_No_of_Dependents,2)
```

```python
min_Total_salary, max_Total_salary = df['Total_salary'].quantile([0.05,0.95])
print(min_Total_salary)
print(round(max_Total_salary,2))
```

```python
df["No_of_Dependents"] = np.where(df["No_of_Dependents"] <  min_No_of_Dependents, min_No_of_Dependents, df["No_of_Dependents"])
df["No_of_Dependents"] = np.where(df["No_of_Dependents"] >  max_No_of_Dependents, max_No_of_Dependents, df["No_of_Dependents"])
df["Total_salary"] = np.where(df["Total_salary"] <  min_Total_salary, min_Total_salary, df["Total_salary"])
df["Total_salary"] = np.where(df["Total_salary"] >  max_Total_salary, max_Total_salary, df["Total_salary"])
```

```python
## now treated the outliers in data
sns.boxplot(data=df,x='No_of_Dependents')
plt.show()
sns.boxplot(data=df,x='Total_salary')
```
